## LSTM - machine learning model
(doesn't need stationary data)

```python
df = df.drop('Temp_Month', axis=1)
df['Date'] = pd.to_datetime(df['Date'])
target = df.iloc[:, 3].values
features = df.iloc[:, 4:].values
```

```python
# Initialize MinMaxScalers for feature and target scaling
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

# Scale the features to range [0, 1]
features_scaled = feature_scaler.fit_transform(features)

# Scale the target and flatten it to a 1D array
target_scaled = target_scaler.fit_transform(target.reshape(-1, 1)).flatten()
```

```python
# Function to create sequences of features and corresponding targets for LSTM input
def create_dataset(features, target, time_steps=1):
    X, y = [], []
    for i in range(len(features) - time_steps):
        X.append(features[i:(i + time_steps)])  # Sequence of time_steps
        y.append(target[i + time_steps])        # Target after the sequence
    return np.array(X), np.array(y)

# Function to define and compile an LSTM model with given hyperparameters
def create_lstm_model(input_shape, n_units, learning_rate, dropout_rate):
    model = Sequential()
    model.add(LSTM(units=n_units, input_shape=input_shape, return_sequences=False))
    model.add(Dense(1))  # Output layer with 1 unit
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='mse', optimizer=optimizer)  # Using mean squared error as the loss function
    return model
```

```python
# Objective function for Optuna to optimize hyperparameters
def objective(trial):
    # Suggest hyperparameters to test
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-2)
    n_units = trial.suggest_int('n_units', 32, 128)
    dropout_rate = trial.suggest_uniform('dropout_rate', 0.0, 0.5)
    time_steps = trial.suggest_int('time_steps', 1, 30)

    # Prepare the dataset using the suggested time_steps
    X, y = create_dataset(features_scaled, target_scaled, time_steps)

    # Split the data into training and validation sets (80/20 split)
    train_size = int(len(X) * 0.8)
    X_train, X_val = X[:train_size], X[train_size:]
    y_train, y_val = y[:train_size], y[train_size:]

    # Define the input shape for the LSTM model
    input_shape = (X_train.shape[1], X_train.shape[2])

    # Build and train the LSTM model
    model = create_lstm_model(input_shape, n_units, learning_rate, dropout_rate)
    model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val), verbose=0)

    # Evaluate the model on the validation set and return the loss
    val_loss = model.evaluate(X_val, y_val, verbose=0)
    return val_loss
```

```python
# Create an Optuna study to minimize the validation loss
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=20)  # Run optimization for 20 trials

# Print the best set of hyperparameters
best_params = study.best_trial.params
print(f"Best Parameters: {best_params}")

# Recreate dataset using the best time_steps found by Optuna
X, y = create_dataset(features_scaled, target_scaled, best_params['time_steps'])

# Train/validation split using the same logic
train_size = int(len(X) * 0.8)
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]

# Define the input shape and build the final model with best parameters
input_shape = (X_train.shape[1], X_train.shape[2])
best_model = create_lstm_model(
    input_shape,
    best_params['n_units'],
    best_params['learning_rate'],
    best_params['dropout_rate']
)

# Train the final model with slightly different batch size
best_model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val), verbose=0)
```

```python
# Model Evaluation
y_pred = best_model.predict(X_val)
y_pred_real = target_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
y_val_real = target_scaler.inverse_transform(y_val.reshape(-1, 1)).flatten()

mse = mean_squared_error(y_val_real, y_pred_real)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_val_real, y_pred_real)
final_loss = best_model.evaluate(X_val, y_val)

print(f"Model Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Final model validation loss: {final_loss:.4f}")
```

```python
# Number of future steps to predict (next 30 months)
future_steps = 30
predictions = []

# Start prediction using the last sequence from the training data
last_data = X[-1]

# Predict step-by-step for future_steps
for i in range(future_steps):
    # Predict the next value using the current sequence
    pred = best_model.predict(last_data.reshape(1, *last_data.shape))[0, 0]
    predictions.append(pred)

    # Shift the sequence one step to the left and add the new prediction at the end
    last_data = np.roll(last_data, shift=-1, axis=0)
    last_data[-1] = np.append(last_data[-1, 1:], pred)  # Update the last timestep

# Inverse transform predictions to original target scale
predictions_original_scale = target_scaler.inverse_transform(np.array(predictions).reshape(-1, 1)).flatten()

# Print each prediction with corresponding step
print("\nPredictions in original scale:")
for i, value in enumerate(predictions_original_scale, start=1):
    print(f"Step {i}: Predicted price = {value:.2f} $/t")
```

```python
# Prepare the x-axis indices for plotting
history_index = df.index  # Existing historical index
future_index = range(len(df), len(df) + future_steps)  # Future steps

# Plotting the historical and predicted data
plt.figure(figsize=(12, 6), dpi=100)
plt.plot(history_index, df.iloc[:, 3], label='Historical Prices', color='blue')  # Plot actual prices
plt.plot(future_index, predictions_original_scale, label='Predicted Future Prices', color='red', linestyle='dashed')  # Plot predictions
plt.title("HRC Price Prediction for Next 30 Days")
plt.ylabel("Price ($/t)")
plt.legend()
plt.xticks([])  # Optional: remove x-axis ticks for cleaner look
plt.grid(True)
plt.show()
```

```python
# Plotting actual vs forecasted in the test set
df['Date']=df['Date'].astype(str)
time_index=df[-len(y_val):]['Date']
plt.figure(figsize=(12,6),dpi=50)
plt.plot(time_index,y_val_real,label='Actual',color='k',linestyle='-')
plt.plot(time_index,y_pred_real,label='Forecasted',color='r',linestyle='-')
plt.title("Actual vs Forecasted HRC Prices")
plt.xlabel("Date")
plt.ylabel("Price ($/t)")
plt.xticks(time_index[::5],rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```
